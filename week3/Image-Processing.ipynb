```python
import cv2
import matplotlib.pyplot as plt
import numpy
```

```python
img_original = cv2.imread('./peppers.png',1)
img_rgb = cv2.cvtColor(img_original,cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img_original,cv2.COLOR_BGR2GRAY)
```

```python
fig = plt.figure(figsize=(10, 7))
rows = 1
columns = 3

fig.add_subplot(rows,columns,1)
plt.title("BGR")
plt.imshow(img_original)
plt.axis("off")


fig.add_subplot(rows,columns,2)
plt.title("RGB")
plt.imshow(img_rgb,)
plt.axis("off")

fig.add_subplot(rows,columns,3)
plt.title("GRAY")
plt.imshow(img_gray,cmap='gray')
plt.axis("off")  
```

```python
# KODLARLA RENK OLUSTURMA 
height,width = 100,100
img_red = np.zeros((height,width,3),dtype = np.uint8)
img_red[:,:,0] = 255
img_blue = np.zeros((height,width,3),dtype = np.uint8)
img_blue[:,:,1] = 255
img_green = np.zeros((height,width,3),dtype = np.uint8)
img_green[:,:,2] = 255
img_yellow = np.zeros((height,width,3),dtype = np.uint8)
img_yellow[:,:,:2] = 255
img_black = np.zeros((height,width,3),dtype = np.uint8)
img_black[:,:,3:] = 255
img_white = np.zeros((height,width,3),dtype = np.uint8)
img_white[:,:,:] = 255

fig = plt.figure(figsize=(10, 7),facecolor='cyan')
rows = 2
columns = 3

fig.add_subplot(rows,columns,1)
plt.title("Red")
plt.imshow(img_red)
plt.axis("off")


fig.add_subplot(rows,columns,2)
plt.title("Blue")
plt.imshow(img_blue,)
plt.axis("off")

fig.add_subplot(rows,columns,3)
plt.title("Green")
plt.imshow(img_green,)
plt.axis("off")


fig.add_subplot(rows,columns,4)
plt.title("Yellow")
plt.imshow(img_yellow,)
plt.axis("off")

fig.add_subplot(rows,columns,5)
plt.title("White")
plt.imshow(img_white,)
plt.axis("off")

fig.add_subplot(rows,columns,6)
plt.imshow(img_black,)
plt.title("Black")
plt.axis("off")
```
