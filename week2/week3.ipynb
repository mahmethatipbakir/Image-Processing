```python

import matplotlib.pyplot as plt
import cv2 
```

```python
my_img = cv2.imread("nebula.jpg")
```

```python
my_img2 = cv2.cvtColor(my_img,cv2.COLOR_BGR2GRAY)
```

```python
my_img2
```

```python
cv2.imshow("image",my_img2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
print(my_img.shape)
print(my_img2.shape)
```

### esitleme

```python
_, my_binary_thresh_img = cv2.threshold(my_img2,thresh = 125 , maxval =255 , type = cv2.THRESH_BINARY)
```

```python
print(_)
```

```python
plt.imshow(my_binary_thresh_img,cmap= "gray")
plt.axis("off")
```

```python
fig = plt.figure(figsize=(10, 7))
rows = 1
columns = 2

fig.add_subplot(rows,columns,1)
plt.imshow(my_img2,cmap = "gray")
plt.axis("off")

fig.add_subplot(rows,columns,2)
plt.imshow(my_binary_thresh_img,cmap = "gray")
plt.axis("off")
```

```python
_, my_binary_thresh_inv_img = cv2.threshold(my_img2,thresh = 125 , maxval =255 , type = cv2.THRESH_BINARY_INV)
```

```python
fig = plt.figure(figsize=(10, 7))
rows = 1
columns = 2

fig.add_subplot(rows,columns,1)
plt.imshow(my_img2,cmap = "gray")
plt.axis("off")

fig.add_subplot(rows,columns,2)
plt.imshow(my_binary_thresh_inv_img,cmap = "gray")
plt.axis("off")
```

```python
fig = plt.figure(figsize=(10, 7))
rows = 1
columns = 3


fig.add_subplot(rows,columns,1)
plt.imshow(my_img2,cmap = "gray")
plt.axis("off")



fig.add_subplot(rows,columns,2)
plt.imshow(my_binary_thresh_inv_img,cmap = "gray")
plt.axis("off")



fig.add_subplot(rows,columns,3)
plt.imshow(my_binary_thresh_img,cmap= "gray")
plt.axis("off")
```

```python
_, my_binary_thresh_TRUNC_img = cv2.threshold(my_img2,thresh = 125 , maxval =255 , type = cv2.THRESH_TRUNC)
```

```python
fig = plt.figure(figsize=(10, 7))
rows = 2
columns = 2

fig.add_subplot(rows,columns,1)
plt.title("main")
plt.imshow(my_img2,cmap = "gray")
plt.axis("off")



fig.add_subplot(rows,columns,2)
plt.title("THRESH_INV")
plt.imshow(my_binary_thresh_inv_img,cmap = "gray")
plt.axis("off")



fig.add_subplot(rows,columns,3)
plt.title("THRESH")
plt.imshow(my_binary_thresh_img,cmap= "gray")
plt.axis("off")


fig.add_subplot(rows,columns,4)
plt.title("TRUNC")
plt.imshow(my_binary_thresh_TRUNC_img,cmap= "gray")
plt.axis("off")
```

```python
_, my_binary_thresh_Tozero_img = cv2.threshold(my_img2,thresh = 125 , maxval =255 , type = cv2.THRESH_TOZERO)
_, my_binary_thresh_Tozero_inv_img = cv2.threshold(my_img2,thresh = 125 , maxval =255 , type = cv2.THRESH_TOZERO_INV)
```

```python
fig = plt.figure(figsize=(10, 7))
rows = 3
columns = 2

fig.add_subplot(rows,columns,1)
plt.title("main")
plt.imshow(my_img2,cmap = "gray")
plt.axis("off")



fig.add_subplot(rows,columns,2)
plt.title("TURNC")
plt.imshow(my_binary_thresh_TRUNC_img,cmap = "gray")
plt.axis("off")



fig.add_subplot(rows,columns,3)
plt.title("THRESH")
plt.imshow(my_binary_thresh_img,cmap= "gray")
plt.axis("off")


fig.add_subplot(rows,columns,4)
plt.title("THRESH INV")
plt.imshow(my_binary_thresh_inv_img,cmap= "gray")
plt.axis("off")

fig.add_subplot(rows,columns,5)
plt.title("tozero")
plt.imshow(my_binary_thresh_Tozero_img,cmap= "gray")
plt.axis("off")


fig.add_subplot(rows,columns,6)
plt.title("tozero inv")
plt.imshow(my_binary_thresh_Tozero_inv_img,cmap= "gray")
plt.axis("off")
```
