```python
import matplotlib.pyplot as plt
import numpy
import cv2
```

```python
foti = cv2.imread("images.jpg")
foti_gray = cv2.cvtColor(foti,cv2.COLOR_BGR2GRAY)
foti.shape
foti_gray.shape
```

```python
plt.imshow(foti,cmap= "gray")
plt.axis("off")
```

## adaptif esitleme islemleri


```python
"""
ADAPTIVE THRESH MEAN C = Adaptif esikleme yontemi turu
THRESH_BINARY =  ADaptif esikleme isleminde kullanilacak normal(baz alinacak) esikleme turu
255 = En buyuk piksel degeri (esiklemeden sonra verilecek en buyuk piksel degeri)
11 = blok buyuklugu (11 satir , 11 sutunluk) bir matris ile goruntu parcalara ayyrilacak ve her parca icerisinde lokal esikleme yapacak
8 = lokal esiklemeden sonra piksel degerinin cikarilacagi sayi
"""

my_adaptive_binary_img = cv2.adaptiveThreshold(foti_gray,255,
                                              cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,8)
```

```python
fig = plt.figure(figsize=(10, 7))
rows = 1
columns = 2

fig.add_subplot(rows,columns,1)
plt.title("main")
plt.imshow(foti_gray,cmap = "gray")
plt.axis("off")



fig.add_subplot(rows,columns,2)
plt.title("adaptive binary")
plt.imshow(my_adaptive_binary_img,cmap = "gray")
plt.axis("off")
```

```python
my_adaptive_binary_inv_img = cv2.adaptiveThreshold(foti_gray,255,
                                              cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY_INV,11,8)
```

```python
fig = plt.figure(figsize=(10, 7))
rows = 1
columns = 3

fig.add_subplot(rows,columns,1)
plt.title("main")
plt.imshow(foti_gray,cmap = "gray")
plt.axis("off")



fig.add_subplot(rows,columns,2)
plt.title("adaptive binary")
plt.imshow(my_adaptive_binary_img,cmap = "gray")
plt.axis("off")


fig.add_subplot(rows,columns,3)
plt.title("adaptive binary")
plt.imshow(my_adaptive_binary_inv_img,cmap = "gray")
plt.axis("off")
```

```python
my_adaptive_gauss_binary_img = cv2.adaptiveThreshold(foti_gray,255,
                                              cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,8)
```

```python
fig = plt.figure(figsize=(10, 7))
rows = 1
columns = 3

fig.add_subplot(rows,columns,1)
plt.title("main")
plt.imshow(foti_gray,cmap = "gray")
plt.axis("off")



fig.add_subplot(rows,columns,2)
plt.title("adaptive binary")
plt.imshow(my_adaptive_binary_img,cmap = "gray")
plt.axis("off")


fig.add_subplot(rows,columns,3)
plt.title("gauss binary")
plt.imshow(my_adaptive_gauss_binary_img,cmap = "gray")
plt.axis("off")
```
