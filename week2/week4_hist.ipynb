```python
import matplotlib.pyplot as plt
import numpy
import cv2

```

```python
foti = cv2.imread("cameraman.jpg")
foti_gray = cv2.cvtColor(foti,cv2.COLOR_BGR2GRAY)
foti.shape
foti_gray.shape
```

```python
plt.imshow(foti,cmap= "gray")
plt.axis("off")
```

```python
histr  = cv2.calcHist([foti_gray],[0],None,[256],[0,256])

plt.plot(histr)
plt.show()
```

```python
foti_gray.shape
```

```python
foti_gray = cv2.imread("cameraman.jpg",0)
foti_eq_gray_hist  = cv2.equalizeHist(foti_gray)

fig = plt.figure(figsize=(10, 7))
rows = 1
columns = 2

fig.add_subplot(rows,columns,1)
plt.title("main")
plt.imshow(foti_gray,cmap = "gray")
plt.axis("off")



fig.add_subplot(rows,columns,2)
plt.title("equalize hist")
plt.imshow(foti_eq_gray_hist,cmap = "gray")
plt.axis("off")
```

```python
hist_org = cv2.calcHist([foti_gray],[0],None,[256],[0,256])
hist_eq = cv2.calcHist([foti_eq_gray_hist],[0],None,[256],[0,256])
```

```python

fig = plt.figure(figsize=(10, 7))
rows = 1
columns = 2

fig.add_subplot(rows,columns,1)
plt.title("main")
plt.plot(hist_org)




fig.add_subplot(rows,columns,2)
plt.title("equalize hist")
plt.plot(hist_eq)
```
